```python
import pandas as pd
df = pd.read_csv('lab 4-5\\ThiTHPT2018.csv')
df1 = df.copy()
```

```python
#thay thế trường toán bằng giá trị trung bình
TBToan = df1['Toan'].mean()
print(TBToan)
df1['Toan'].fillna(TBToan, inplace = True)
#df1['Toan']
```

```python
#thay thế giá trị thiếu trường văn bằng giá trị thường xuyên
print(df1['Van'].mode())
df1['Van'].fillna(df1['Van'].mode())
```

```python
#thay giá trị thiếu cho trường anh văn
#giatritxcuatinh = df1.groupby('Ten Tinh')['Anh'].apply(lambda x: x.mode())
df1['Anh'] = df1['Anh'].fillna(df1.groupby('Ten Tinh')['Anh'].apply(lambda x: x.mean()))
#df.groupby('Ten Tinh')['Anh'].fillna(....)
#print(giatritxcuatinh)
df1['Anh']
```

```python
#thay giá trị không thi
df1.fillna(value={'Ly': 'khongthi', 'Hoa': 'khongthi', 'Sinh': 'khongthi',
                  'Su': 'khongthi', 'Dia': 'khongthi','GDCD': 'khongthi',
                  'KhoiA': 'khongthi', 'KhoiB': 'khongthi', 'KhoiC': 'khongthi',
                  'KhoiD': 'khongthi', 'KhoiA1': 'khongthi'}, inplace=True)
```

```python
#kiemtratyle[kiemtratyle > 0.5]
kiemtratyle = df1.isnull().mean()
#xóa những cột có giá trị thiếu trên 50%
df1.drop(kiemtratyle[kiemtratyle > 0.5].index[1:len(df1.columns)-3], axis=1)
```

```python
#điền giá trị 0 cho những ô bị thiếu
df1[4:].fillna(0)
```

```python
#điền giá trị NaN cho những ô thiếu
df1[4:].fillna('NaN')
```

```python
#<20%
dulieuthieu = df1.isnull().mean()
#print(dulieuthieu)
# Xác định các cột có tỷ lệ giá trị thiếu dưới 20%
timcot = dulieuthieu[dulieuthieu < 0.2].index
#print(timcot)
for col in timcot:
    if df1[col].dtype == 'object':
        df1[col].fillna(df1[col].mode()[0], inplace=True)
    else:
        df1[col].fillna(df1[col].mean(), inplace=True)
```

```python
#chuẩn hóa
from scipy.stats import zscore
df1.dropna()
df1_data_Hoa = df['Hoa']
df1_data_Ly = df['Ly']
df1_data_Dia = df['Dia']

z_score_hoa=zscore(df1_data_Hoa)
z_score_Ly=zscore(df1_data_Ly)
z_score_Ly=zscore(df1_data_Dia)

z_score_hoa
```
