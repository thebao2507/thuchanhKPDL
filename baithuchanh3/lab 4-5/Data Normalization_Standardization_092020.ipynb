```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt # for plotting graphs
from sklearn.preprocessing import MinMaxScaler #for scaling data
from sklearn.cluster import KMeans # For clustering by using K-Means algorithm
```

```python
#Loading dataset
import pandas as pd
data=pd.read_csv('EmployeeSalary.csv')
data
```

```python
data.describe()
```

```python
data.plot(kind='scatter', x='WorkingYears',y='Salary')
plt.show()
```

```python
#Drop ID column, we don't use this column
df=data.drop(['ID'], axis=1)
```

```python
df
```

```python
df.plot(kind='box')
plt.title('Original data')
plt.show()
```

```python
df.corr()
```

```python
#Scaling the dataset (Normalization): x_mms=(x-x_min)/(x_max- x_min)
mms=MinMaxScaler()
mms.fit(df)
data_mms=mms.transform(df)
```

```python
data_mms
```

```python
df
```

```python
#Convert to Dataframe
data_mms=pd.DataFrame(data_mms, columns=['WorkingYears','Salary'])
data_mms
```

```python
#Replotting dataset after scalling
data_mms.plot(kind='scatter', x='WorkingYears',y='Salary')
plt.show()
```

```python
df.plot(kind='box')
plt.title('Original data')
plt.show()
###
data_mms.plot(kind='box')
plt.title('data after max_min normalization')
plt.show()
```

```python
data_mms.corr()
```

```python
#Standardization: x_stand=(x-mean(x))/Standard Deviation
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data_stand=sc.fit_transform(df)
```

```python
data_stand
```

```python
#Convert to Dataframe
data_stand=pd.DataFrame(data_stand, columns=['WorkingYears','Salary'])
data_stand
```

```python
#Replotting dataset after standardization
data_stand.plot(kind='scatter', x='WorkingYears',y='Salary')
plt.show()
```

```python
df.plot(kind='box')
plt.title('Original data')
plt.show()
###
data_mms.plot(kind='box')
plt.title('data after max_min normalization')
plt.show()
###
data_stand.plot(kind='box')
plt.title('data after standardization')
plt.show()
```
