```python
# values for Gender: Male:1/Female:0
df['Gender_Cat'] = df['Gender'].map({'Male':1, 'Female':0})
# values for EverseaProject: Yes:1/No:0
df['OverseaProject_Cat'] = df['OverseaProject'].map({'Yes':1, 'No':0})
```

```python
#Delete columns
df.drop('Gender', axis=1, inplace=True)
df.drop('OverseaProject', axis=1, inplace=True)
```

```python
# Convert categorical variables to dummies
df_dummies = pd.get_dummies(data=df, columns = ['Skill'])
df_dummies.head()
```

```python
"""
Discretization
Discretization, also known as quantization or binning, 
divides a continuous feature into a pre-specified number of categories (bins), 
and thus makes the data discrete.

"""
```

```python
#Discretization using cut method of pandas package
cut_labels = ['Low','Medium','High'] 
cut_bins = [0, 800, 1200, 2000] #0: min, 2000: max
df['Sal_Cat'] = pd.cut(df['Salary'], bins=cut_bins, labels=cut_labels)
```

```python
df
```

```python
#Discretization with bin parameter
df['Sal_Category']=pd.cut(df['Salary'],3,labels=['Low','Medium','High'])
df.head(10)
```

```python
df['Working_Level']=pd.cut(df['WorkingYear'],4,labels=['Internship','Fresher','Junior', 'Senior'])
df.head(10)
```

```python
#Discretization using cut method of pandas package
cut_TrainedYear_labels = ['College','University','Master_PhD'] 
cut_TrainedYear_bins = [0, 4, 6, 10] #0: min, 2000: max
df['TrainedYear_Cat'] = pd.cut(df['TrainedYear'], bins=cut_TrainedYear_bins, 
                               labels=cut_TrainedYear_labels)
```

```python
df
```

```python
#Discretization using qcut method of pandas package
bin_labels=['Internship','Fresher','Junior', 'Senior']
df['Working_Level_qcut']=pd.qcut(df['WorkingYear'],q=[0,0.25,0.5,0.75,1],labels=bin_labels)
df.head(10)
```
