```python

import pandas as pd
import numpy as np
data=pd.read_excel('JOB_INFO.xlsx')
```

```python
data
```

```python
#Show null value of columns
data.isnull().sum()
```

```python
#Show all row has null value
mask=False
for col in data.columns: mask = mask | data[col].isnull()
datanulls = data[mask]
datanulls
```

```python
#Sum of null values
data.isnull().sum().sum()
```

```python
#Show columns has null value
null_columns=data.columns[data.isnull().any()]
data[null_columns].isnull().sum()
```

```python
#1.Deleting Row has null value (not recommend)
data1=data.copy()
data1.dropna(inplace=True)
data1.isnull().sum()
```

```python
data1.shape
```

```python
#2.Replacing Null values With Mean/Median/Mode
data2=data.copy()
```

```python
data2.describe()
```

```python
data2
```

```python
#Update Specific row and column
#In this case change Age of row index 2 (the third row) into 30
#data2.at[2,'Age']= 30
```

```python
#Replacing null value with mean of Age Attribute
data2['Age'].fillna(data2['Age'].mean(), inplace=True)
```

```python
data2
```

```python
#Replacing null value with mean of all quantitative columns
data2= data2.fillna(data2.mean())
```

```python
data2
```

```python
data2.isnull().sum()
```

```python
data2
```

```python
#Fille null categorical value with Mode
data2['Gender'].fillna(data2['Gender'].mode()[0], inplace=True)
```

```python
data2['Gender'].mode()
```

```python
data2
```

```python
data2.isnull().sum()
```

```python
#Fille null value with 'Unknown'
data2['Dept'].fillna('Unknown', inplace=True)
```
