```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_excel('lab 4-5\\marketing.xlsx')
df
```

```python
scaler = MinMaxScaler()

# Thực hiện chuẩn hóa trên dataframe
df_norm = scaler.fit_transform(df)

# Chuyển đổi kết quả từ ndarray thành dataframe
df_norm = pd.DataFrame(df_norm, columns=df.columns)

# In kết quả
print(df_norm)
```

```python
print('Min max scaling')
from sklearn import preprocessing as pp
mms = pp.MinMaxScaler()
data_mms = mms.fit_transform(df)
print(data_mms)
```
