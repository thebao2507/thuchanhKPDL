```python
import pandas as pd
import seaborn as sns
df = pd.read_csv("lab 2-3\\ThiTHPT2018.csv")
df
```

```python
#tính trung bình môn toán
trungbinhtoan = df['Toan'].mean()
trungbinhtoan
```

```python
#tính trung bình môn văn
trungbinhvan = df['Van'].mean()
trungbinhvan
```

```python
#tính trung bình môn anh
trungbinhanh = df['Anh'].mean()
trungbinhanh
```

```python
#tính 
dolechToan = df['Toan'].std()
dolechToan
```

```python
#tính trung vị
trungviToan = df['Toan'].median()
trungviToan
```

```python
sns.boxplot(x=df['Toan'])
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=(20,10))
# Vẽ biểu đồ histogram cho trường các môn học
df['Toan'].hist(bins=70, alpha=0.5, ax=ax, label='Toan')
df['Van'].hist(bins=70, alpha=0.5, ax=ax, label='Van')
df['Anh'].hist(bins=70, alpha=0.5, ax=ax, label='Anh')
df['Ly'].hist(bins=70, alpha=0.5, ax=ax, label='Ly')
df['Hoa'].hist(bins=70, alpha=0.5, ax=ax, label='Hoa')
df['Sinh'].hist(bins=70, alpha=0.5, ax=ax, label='Sinh')
df['Su'].hist(bins=70, alpha=0.5, ax=ax, label='Su')
df['Dia'].hist(bins=70, alpha=0.5, ax=ax, label='Dia')
df['GDCD'].hist(bins=70, alpha=0.5, ax=ax, label='GDCD')

# Đặt tiêu đề cho biểu đồ
ax.set_title("thống kê điểm môn học")

# Thêm nhãn cho trục hoành
ax.set_xlabel("điểm số các môn")

# Thêm nhãn cho trục tung
ax.set_ylabel("Số lượng học sinh")

# Hiển thị legend
ax.legend()

# Hiển thị biểu đồ
plt.show()

```
