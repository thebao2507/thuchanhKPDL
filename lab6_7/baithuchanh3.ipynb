```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('Market_Basket_AR.csv')
df
```

```python
df = df.applymap(lambda x: 0 if pd.isnull(x) else 1)
df
```

```python
frequent_itemsets_60 = apriori(df.astype('bool'), min_support=0.6, use_colnames=True)
frequent_itemsets_60
```

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets_60, metric="confidence", min_threshold=0.85)
rules
```
