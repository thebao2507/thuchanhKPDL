```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
df = pd.read_csv('105OK.csv')
df
```

```python
df_10 = df.iloc[:, :10]
df_10
```

```python
frequent_itemsets_40 = apriori(df_10, min_support=0.4, use_colnames=True)
frequent_itemsets_40
```

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets_40, metric="confidence", min_threshold=0.6)
rules
```
